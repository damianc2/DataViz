```python
import numpy as np
import pandas as pd
```

```python
# https://stackoverflow.com/questions/48067514/utf-8-codec-cant-decode-byte-0xa0-in-position-4276-invalid-start-byte
# source for utf-8 error (encoding windows 1252)
df = pd.read_csv("store_data.csv", encoding='windows-1252')
df
```

```python
# https://coderzcolumn.com/tutorials/data-science/interactive-plotting-in-python-jupyter-notebook-using-bqplot
# source for guidance on using bqplot
from bqplot import pyplot as plt

quan_df = df.groupby(by='Quantity').mean().reset_index()
quan_df["Quantity"] = ["{}".format(cat) for cat in quan_df.Quantity]
quan_df
```

```python
from bqplot import Tooltip

fig = plt.figure(title = "Average Discount per Quantity Amount")

fig.layout.height = '400px'
fig.layout.width = "600px"

bar_chart  = plt.bar(x = quan_df.Quantity.tolist(), y= quan_df["Discount"])

bar_chart.colors = ["mediumblue"]

bar_chart.tooltip = Tooltip(fields=["x", "y"], labels=["Quantity", "Average Discount"])

plt.xlabel("Quantity")
plt.ylabel("Average Discount")

plt.show()
```

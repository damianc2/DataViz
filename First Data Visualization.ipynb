```python
import pandas as pd
df = pd.read_csv('Raisin_Dataset.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['label'], df['minor_axis_length'])
plt.title('Minor Axis Length of Both Raisin Types')
plt.xlabel('Label')
plt.ylabel('Minor Axis Length')
```

```python
plt.bar(df['label'], df['major_axis_length'])
plt.title('Major Axis Length of Both Raisin Types')
plt.xlabel('Label')
plt.ylabel('Major Axis Length')
```

```python
plt.bar(df['label'], df['perimeter'])
plt.title('Perimeter of Both Raisin Types')
plt.xlabel('Label')
plt.ylabel('Perimeter')
```

```python
plt.scatter(df['area'], df['major_axis_length'])
plt.title('Relationship between Major Axis Length and Area')
plt.xlabel('Area')
plt.ylabel('Major Axis Length')
```

```python
plt.scatter(df['area'], df['minor_axis_length'])
plt.title('Relationship between Minor Axis Length and Area')
plt.xlabel('Area')
plt.ylabel('Minor Axis Length')
```

```python
plt.scatter(df['eccentricity'], df['extent'])
plt.title('Relationship between Eccentricity and Extent')
plt.xlabel('Eccentricity')
plt.ylabel('Extent')
```
